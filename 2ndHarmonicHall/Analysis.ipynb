```python
import matplotlib.pyplot as plt
import numpy as np

# -------------------------
# Data
# -------------------------

data = {
    "MgO": {
        "t_FM": np.array([2.5, 3, 10]),
        "eff": np.array([
            -3.867070218551377e-06,
            -2.0792681953063495e-05,
            2.211135403377806e-05
        ]),
        "err": np.array([
            4.441507771060816e-06,
            1.7055861783352198e-05,
            1.1824191945581402e-05
        ])
    },
    "Pt5": {
        "t_FM": np.array([2, 4, 8, 10]),
        "eff": np.array([
            3.103788632102104e-03,
            1.4959204114370242e-03,
            6.236997633753406e-04,
            3.2685389165707987e-04
        ]),
        "err": np.array([
            2.2763789527116393e-05,
            5.185278831460807e-05,
            1.7790666562171421e-04,
            1.9731025301071193e-05
        ])
    },
    "Pt15": {
        "t_FM": np.array([2, 4, 10]),
        "eff": np.array([
            1.2208327602509456e-04,
            4.277488600238566e-05,
            2.226381603041635e-04
        ]),
        "err": np.array([
            6.9539389098157385e-06,
            1.6034553754494014e-05,
            7.548112147650234e-05
        ])
    }
}

# -------------------------
# Plot
# -------------------------

plt.figure(figsize=(7, 5))

markers = {"MgO": "o", "Pt5": "s", "Pt15": "^"}

for label, d in data.items():
    plt.errorbar(
        d["t_FM"],
        d["eff"],
        yerr=d["err"],
        fmt=markers[label],
        capsize=4,
        markersize=7,
        linestyle="None",
        label=label
    )

plt.axhline(0, color="gray", linestyle="--", linewidth=1)

plt.xlabel(r"$t_{\mathrm{FM}}$ (nm)")
plt.ylabel(r"DL efficiency")
plt.title("DL efficiency vs FM thickness")
plt.legend()
plt.tight_layout()
plt.show()
```
