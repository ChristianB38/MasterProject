```python
import matplotlib.pyplot as plt
t = [3, 7, 9, 13, 18]

xi_angular = [0.066, 0.033, 0.016, -0.032, -0.12]
xi_first_sol = [0, 0.12, 0.081, 0.085, 0.11]
xi_second_sol = [0.080, 0.069, 0.063, 0.057, 0.058]

plt.figure(figsize=(8, 6))
plt.plot(t, xi_angular, 'o-', label=r'Conventional method')
plt.plot(t[1:], xi_first_sol[1:], 's-', label=r'$\xi_{FMR}$ with first solution AHE-PHE')
plt.plot(t, xi_second_sol, '^-', label=r'$\xi_{FMR}$ with second solution AHE-AMR')
plt.legend()
plt.title(r'$\xi_{FMR}$ vs Pt thickness')
plt.grid(True)
plt.show()
```
