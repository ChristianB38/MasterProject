```python
import matplotlib.pyplot as plt
import numpy as np
t = [3, 7, 9, 13, 18]

xi_angular = [0.066, 0.030, 0.016, -0.059, -0.14]
xi_first_sol = [0, 0.12, 0.081, 0.085, 0.106]
xi_second_sol = [0.073, 0.070, 0.053, 0.056, 0.058]

plt.figure(figsize=(8, 6))
plt.plot(t, xi_angular, 'o-', label=r'$\xi_{FMR} Angular$')
plt.plot(t[1:], xi_first_sol[1:], 's-', label=r'$\xi_{FMR} First solution$')
plt.plot(t, xi_second_sol, '^-', label=r'$\xi_{FMR} Second solution$')
plt.legend()

plt.grid(True)

plt.show()


xi_first_sol_avg = np.mean(xi_first_sol[1:])


e = 1.602176634e-19     # elementary charge [C]
hbar = 1.055e-34
current = ((2 * e) / (hbar * xi_first_sol_avg)) * 5e-9 * 1.2 * 2.5
print(f"Current density: {current:.2e} A/m^2")
```
